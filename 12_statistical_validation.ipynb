Test 1 - Does average order value differ across customer segments?

    H0: mean AOV is equal across all segments
    H1: at least one segment has a different mean AOV

    Test: one way ANOVA - continuous outcome, 3+ independent groups


1 Average Order Value differences across customer segments

```python
#get all the tables:
from db_setup import conn, currencyexchange, customer, date, product, sales, store
#instead of saving csv files in sql and then importing one by one
import pandas as pd
import numpy as np

customer_summary=pd.read_parquet(r"C:\Users\amyre\Desktop\Luke Barousse SQL Course\SQL + Contoso\saved_files\customer_summary.parquet")
order_enriched=pd.read_parquet(r"C:\Users\amyre\Desktop\Luke Barousse SQL Course\SQL + Contoso\saved_files\order_enriched.parquet")
customer_segments=pd.read_parquet(r"C:\Users\amyre\Desktop\Luke Barousse SQL Course\SQL + Contoso\saved_files\customer_segments.parquet")
```

```python
from scipy import stats

merged_test1 = order_enriched.merge(
    customer_segments,
    on="customerkey",
    how="inner"
)

filtered = merged_test1[merged_test1["order_revenue"].notna()]

result_test1 = filtered[["combined_segment", "order_revenue"]]

groups = [
    g["order_revenue"].values
    for _, g in result_test1.groupby("combined_segment")
]

f_stat, p_value = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("p-value:", p_value)

```

Test 2 - Does delivery time differs across countries?

Delivery time differences across countries

```python
delivery_data = order_enriched.dropna(subset=["delivery_days"])

groups = [
    g["delivery_days"].values
    for _, g in delivery_data.groupby("customer_country")
]

f_stat, p_value = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("p-value:", p_value)
```
